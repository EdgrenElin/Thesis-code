```python

# Put together for easier handling
dataframes = {
    "train": [train1, train2, train3, train4, train5, train6, train7, train8, train9, train10],
    "val": [val1, val2, val3, val4, val5, val6, val7, val8, val9, val10]
}
```

```python

# Get residuals and L1 
residuals = {"train": [], "val": []}
L1_norms = {"train": [], "val": []}

for key, df_list in dataframes.items():
    for df in df_list:
        df_residuals = df['labels'] - df['preds']
        
        L1 = np.abs(df_residuals)
        
        residuals[key].append(df_residuals)
        L1_norms[key].append(L1)

```

######################################################################################################
- This code looks at the correlation between cosine similarity for kNN in embeddings space and the predictive error for the embeddings.
######################################################################################################

```python


fig=plt.figure(figsize=(12, 9))
colors_plot = ['red', 'orange', 'green', 'blue', 'black', 'maroon', 'cyan', 'magenta', 'darkviolet', 'chartreuse']

for i in range(10):
    emb_val = np.asarray(dataframes['val'][i].CLS_embeddings.tolist(), dtype=np.float32)
    emb_train = np.asarray(dataframes['train'][i].CLS_embeddings.tolist(), dtype=np.float32)
    error_fold = L1_norms['val'][i]
    
    neighbor_sizes = np.linspace(1, len(emb_train), num=20, dtype=int)
    neighbor_sizes = np.unique(neighbor_sizes)
    
    correlations_k_nearest_median = {}
    correlations_k_nearest_25 = {}
    correlations_k_nearest_75 = {}

    for k in neighbor_sizes:
        cosine_k_median, cosine_k_25, cosine_k_75, error_k = [], [], [], []
        
        for j in range(len(emb_val)):
            current_embedding = emb_val[j].reshape(1, -1)
            cossim = cosine_similarity(emb_train, current_embedding).flatten()
            top_k_idx = np.argsort(-cossim)[:k]
            
            cosine_k_median.append(np.median(cossim[top_k_idx]))
            cosine_k_25.append(np.percentile(cossim[top_k_idx], 25))
            cosine_k_75.append(np.percentile(cossim[top_k_idx], 75))
            error_k.append(error_fold[j])
        
        correlations_k_nearest_median[k] = scipy.stats.spearmanr(cosine_k_median, error_k)[0]
        correlations_k_nearest_25[k] = scipy.stats.spearmanr(cosine_k_25, error_k)[0]
        correlations_k_nearest_75[k] = scipy.stats.spearmanr(cosine_k_75, error_k)[0]

    k_values_for_plot = neighbor_sizes.tolist()
    correlations_for_plot = [correlations_k_nearest_median[k] for k in neighbor_sizes]
    correlations_25_for_plot = [correlations_k_nearest_25[k] for k in neighbor_sizes]
    correlations_75_for_plot = [correlations_k_nearest_75[k] for k in neighbor_sizes]

    # Plot with fill_between for 25th-75th percentile range
    plt.fill_between(k_values_for_plot, correlations_25_for_plot, correlations_75_for_plot, color=colors_plot[i], alpha=0.3)
    plt.plot(k_values_for_plot, correlations_for_plot, 'o-', color=colors_plot[i], label=f"Fold {i+1}", markersize=4)

plt.legend(title="Folds", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.title("Spearman Correlation between Cosine Similarity of the k-NN and L1 Error")
plt.xlabel("k-Nearest Neighbors count")
plt.ylabel("Spearman Correlation")
plt.show()
plt.savefig("corrBig_fig.pdf",format='pdf', bbox_inches='tight')
```
